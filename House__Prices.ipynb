```python
most_common_Lot = data['LotFrontage'].mode()[0]
data['LotFrontage'] = data['LotFrontage'].fillna(most_common_Lot)

most_common_Alley = data['Alley'].mode()[0]
data['Alley'] = data['Alley'].fillna(most_common_Alley)

most_common_MasVnrType = data['MasVnrType'].mode()[0]
data['MasVnrType'] = data['MasVnrType'].fillna(most_common_MasVnrType)

most_common_MasVnrArea = data['MasVnrArea'].mode()[0]
data['MasVnrArea'] = data['MasVnrArea'].fillna(most_common_MasVnrArea)

most_common_BsmtQual = data['BsmtQual'].mode()[0]
data['BsmtQual'] = data['BsmtQual'].fillna(most_common_BsmtQual)

most_common_BsmtCond = data['BsmtCond'].mode()[0]
data['BsmtCond'] = data['BsmtCond'].fillna(most_common_BsmtCond)

most_common_BsmtExposure = data['BsmtExposure'].mode()[0]
data['BsmtExposure'] = data['BsmtExposure'].fillna(most_common_BsmtExposure)

most_common_BsmtFinType1 = data['BsmtFinType1'].mode()[0]
data['BsmtFinType1'] = data['BsmtFinType1'].fillna(most_common_BsmtFinType1)

most_common_BsmtFinType2 = data['BsmtFinType2'].mode()[0]
data['BsmtFinType2'] = data['BsmtFinType2'].fillna(most_common_BsmtFinType2)

most_common_Electrical = data['Electrical'].mode()[0]
data['Electrical'] = data['Electrical'].fillna(most_common_Electrical)

most_common_FireplaceQu = data['FireplaceQu'].mode()[0]
data['FireplaceQu'] = data['FireplaceQu'].fillna(most_common_FireplaceQu)

most_common_GarageType = data['GarageType'].mode()[0]
data['GarageType'] = data['GarageType'].fillna(most_common_GarageType)

most_common_GarageYrBlt = data['GarageYrBlt'].mode()[0]
data['GarageYrBlt'] = data['GarageYrBlt'].fillna(most_common_GarageYrBlt)

most_common_GarageFinish = data['GarageFinish'].mode()[0]
data['GarageFinish'] = data['GarageFinish'].fillna(most_common_GarageFinish)

most_common_GarageQual = data['GarageQual'].mode()[0]
data['GarageQual'] = data['GarageQual'].fillna(most_common_GarageQual)

most_common_GarageCond = data['GarageCond'].mode()[0]
data['GarageCond'] = data['GarageCond'].fillna(most_common_GarageCond)

most_common_PoolQC = data['PoolQC'].mode()[0]
data['PoolQC'] = data['PoolQC'].fillna(most_common_PoolQC)

most_common_Fence = data['Fence'].mode()[0]
data['Fence'] = data['Fence'].fillna(most_common_Fence)

most_common_MiscFeature = data['MiscFeature'].mode()[0]
data['MiscFeature'] = data['MiscFeature'].fillna(most_common_MiscFeature)
```

```python
missing_features = data.columns[data.isna().any()].tolist()
print("Eksik veri içeren feature'lar:\n", missing_features)
```

```python
data = pd.get_dummies(data,drop_first=True)
data.head()
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

X = data.drop('SalePrice', axis=1)
Y = data['SalePrice']

selector = SelectKBest(score_func=f_regression, k=5)
X_selected = selector.fit_transform(X, Y)

selected_indices = selector.get_support(indices=True)

selected_features = X.columns[selected_indices]

print("Selected Features:")
print(selected_features)
```

```python
from sklearn.model_selection import train_test_split

X = data[['OverallQual', 'TotalBsmtSF', 'GrLivArea', 'GarageCars', 'GarageArea']]
y = data['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)

print("Eğitim Doğruluğu:", train_accuracy)
print("Test Doğruluğu:", test_accuracy)
```
